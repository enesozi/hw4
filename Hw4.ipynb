Suppose we have a $3×3$  matrix and wish to introduce zeros by left- and/or right- multiplications by unitary matrices $Q_j$ , such as Householder reﬂectors or Givens rotations. Consider the following three matrix structures:
$$
\text{a)}
\begin{bmatrix}
    x & x & 0 \\
    0 & x & x \\
    0 & 0 & x
\end{bmatrix}\hspace{1cm}
\text{b)}
\begin{bmatrix}
    x & x & 0 \\
    x & 0 & x \\
    0 & x & x
\end{bmatrix}\hspace{1cm}
\text{c)}
\begin{bmatrix}
    x & x & 0 \\
    0 & 0 & x \\
    0 & 0 & x
\end{bmatrix}
$$

For each one, decide which of the following situations holds, and justify your claim.

(i) Can be obtained by a sequence of left-multiplications by matrices $Q_j$ ;

(ii) Not (i), but can be obtained by a sequence of left- and right-multiplications by matrices $Q_j$;

(iii) Cannot be obtained by a sequence of left- and right-multiplications by matrices $Q_j$.


Let's start with the matrix in C.

### c)
Matrix is 2-rank, i.e singular. Since transformation between general matrix and the singular matrix is not possible with multiplication of unitary matrices with general matrix. Therefore answer is  (iii). 

### a)
For this matrix, we multiply diﬀerent $Q_j$ on the left and right. Then we multiply with $Q_1$ on left to the matrix A to make a *Householder* reﬂector :
$$
\text{A} = 
\begin{bmatrix}
    x & x & x \\
    x & x & x \\
    x & x & x
\end{bmatrix}
$$
$$
A_1 = Q_1A =
\begin{bmatrix}
    x & x & x \\
    0 & x & x \\
    0 & x & x
\end{bmatrix}
$$

Then we apply *Givens* rotation on the right to $A_1$ s.t $A_1(1,2:3) = [x,0]$.

$$
A_2 = A_1Q_2 \\
A_2 = 
\begin{bmatrix}
    x & x & 0 \\
    0 & x & x \\
    0 & x & x
\end{bmatrix}
$$

Then we apply Householder reflector on to left $A_2$ s.t $A_2(2,2:3) = [x,0]'$

$$
A_3 = Q_3A_2 \\
A_3 = 
\begin{bmatrix}
    x & x & 0 \\
    0 & x & x \\
    0 & 0 & x
\end{bmatrix}
$$

Since we can obtain the desired matrix with multiplication of unitary matrices on a general matrix, the answer is (ii).
### b)

This matrix can be obtained by permutation matrix on to the matrix we obtained in (a).
$$
P = 
\begin{bmatrix}
    1 & 0 & 0 \\
    0 & 0 & 1 \\
    0 & 1 & 0
\end{bmatrix}
$$
$$
A_4 = PA_3 = PQ_3Q_1AQ_2
$$

with an extra zero we can obtain matrix in (b).

Therefore the anser is (ii).
